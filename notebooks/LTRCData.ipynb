```python
import pandas as pd
```

```python
train_data = pd.read_csv('../data/ltrc/Hatespeech_train.csv')
test_data = pd.read_csv('../data/ltrc/Hatespeech_test.csv')
val_data = pd.read_csv('../data/ltrc/Hatespeech_valid .csv')
```

```python
train_data['text'] = train_data['Sentence']
train_data['label_yn'] = train_data['Hate-Speech'].apply(lambda x: 0 if x == 'no' else 1)
test_data['text'] = test_data['Sentence']
test_data['label_yn'] = test_data['Hate-Speech'].apply(lambda x: 0 if x == 'no' else 1)
val_data['text'] = val_data['Sentence']
val_data['label_yn'] = val_data['Hate-Speech'].apply(lambda x: 0 if x == 'no' else 1)
train_data['uid'] = ['ltrc_tel_train_' + str(i) for i in range(len(train_data))]
test_data['uid'] = ['ltrc_tel_test_' + str(i) for i in range(len(test_data))]
val_data['uid'] = ['ltrc_tel_val_' + str(i) for i in range(len(val_data))]
train_data.head()
```

```python
val_data[['uid', 'text', 'label_yn']]
```

```python

output_test_path = 'ltrc_tel_test.csv'
output_train_path = 'ltrc_tel_train.csv'
output_val_path = 'ltrc_tel_val.csv'

ltrc_train = train_data[['uid', 'text', 'label_yn']]
ltrc_test = test_data[['uid', 'text', 'label_yn']]
ltrc_val = val_data[['uid', 'text', 'label_yn']]

ltrc_val.to_csv(output_val_path, index=False)
ltrc_train.to_csv(output_train_path, index=False)
ltrc_test.to_csv(output_test_path, index=False)

ltrc = pd.concat([ltrc_train, ltrc_test, ltrc_val], ignore_index=True)
ltrc.to_csv('ltrc_tel.csv', index=False)
```
