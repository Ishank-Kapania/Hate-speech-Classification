```python
def get_labelled_dataset(train, test, val):
    train_h = train[train['Labels Set'].apply(lambda x: 'hate' in x or 'offensive' in x)]
    test_h = test[test['Labels Set'].apply(lambda x: 'hate' in x or 'offensive' in x)]
    val_h = val[val['Labels Set'].apply(lambda x: 'hate' in x or 'offensive' in x)]

    train_nh = train[train['Labels Set'].apply(lambda x: 'non-hostile' in x)]
    test_nh = test[test['Labels Set'].apply(lambda x: 'non-hostile' in x)]
    val_nh = val[val['Labels Set'].apply(lambda x: 'non-hostile' in x)]

    train_h['label_yn'] = 1
    train_nh['label_yn'] = 0
    test_h['label_yn'] = 1
    test_nh['label_yn'] = 0
    val_h['label_yn'] = 1
    val_nh['label_yn'] = 0

    train_h['uid'] = train_h['Unique ID'].apply(lambda x: generate_unique_id('constai_hi_train', x))
    test_h['uid'] = test_h['Unique ID'].apply(lambda x: generate_unique_id('constai_hi_test', x))
    val_h['uid'] = val_h['Unique ID'].apply(lambda x: generate_unique_id('constai_hi_val', x))
    train_nh['uid'] = train_nh['Unique ID'].apply(lambda x: generate_unique_id('constai_hi_train', x))
    test_nh['uid'] = test_nh['Unique ID'].apply(lambda x: generate_unique_id('constai_hi_test', x))
    val_nh['uid'] = val_nh['Unique ID'].apply(lambda x: generate_unique_id('constai_hi_val', x))

    train_h.index = train_h.uid
    test_h.index = test_h.uid
    val_h.index = val_h.uid
    train_nh.index = train_nh.uid
    test_nh.index = test_nh.uid
    val_nh.index = val_nh.uid

    train_df = pd.concat([train_h, train_nh], axis=0)
    test_df = pd.concat([test_h, test_nh], axis=0)
    val_df = pd.concat([val_h, val_nh], axis=0)

    train_df['text'] = train_df.Post
    test_df['text'] = test_df.Post
    val_df['text'] = val_df.Post

    return train_df[['text', 'label_yn']], test_df[['text', 'label_yn']], val_df[['text', 'label_yn']] 
```

```python
cds = get_labelled_dataset(cosaidf, cosaidf2, cosaidf3)
```
